```python
# import all

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import pairwise_distances
from scipy.cluster import hierarchy


%matplotlib inline
import numpy as np
import scipy.io as spio
import matplotlib.pyplot as plt
import matplotlib.image as image
from skimage import io
```

```python
# prepare data
matdata = spio.loadmat('data2.mat')
print(matdata.keys())
dataset = pd.DataFrame(matdata['X'], columns=["x0", "x1"])
dataset
```

```python
# visualize the data

dataset.plot.scatter(x='x0', y='x1', c='DarkBlue')
```

### Exercício 1
#### a) Executar  o  algoritmo K-means para  obter  3  clusters  das  amostras presentes no arquivo “data2.mat”, conforme Figura 1. 

```python
# create k-means

kmeans = KMeans(n_clusters=3, random_state=1).fit(dataset)
split_sep = np.random.rand(len(dataset)) < 0.8
kmeans.inertia_
```

#### b) Agora ajuste o algoritmo K-Means para diferentes números de clusters (1 a 20).  
#### c) Para cada modelo, armazene o número de clusters e o valor da inércia (custo)

```python
# store the models

kmeans_models = []
for i in range(20):
    kmeans_models.append(KMeans(n_clusters=i+1, random_state=1).fit(dataset))
```

#### d) Plote  a Inércia  (custo)  versus  Nº  de  clusters.  Parece  haver  um  número ideal de cluster (método do cotovelo -elbow?

```python
fig, ax = plt.subplots()

ax.set(title='Inertia x N. Clusters', xlabel='clusters', ylabel='inertia')
ax.plot([c for (c, km) in enumerate(kmeans_models)], 
        [km.inertia_ for km in kmeans_models])
```

#### e) Ajuste    um    modelo de    agrupamento    aglomerativo hierárquico (Hierarchical  Agglomerative  Clustering) com  três  clusters(escolha  e teste diferentes tipos de linkage).

```python
# Agglomerative Clustering
# The linkage criterion determines which distance to use between sets of observation.

linkages = ['ward', 'complete', 'average', 'single']
agc_models = {}

for linkage in linkages:
    agc_models[linkage] = AgglomerativeClustering(n_clusters=3, linkage=linkage).fit(dataset)
```

#### f) Compare os resultados com os obtidos pelo k-means

```python
# Visualising the clusters
def plot_kmean_cluster(kmeans, title, ax):
    y_kmeans = kmeans.predict(dataset)
    centers = kmeans.cluster_centers_
    ax.scatter(dataset['x0'], dataset['x1'], c=y_kmeans, s=50, cmap='viridis')
    ax.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
    ax.set_title(title)
    
def plot_agc_cluster(agc, title, ax):
    y_agc = agc.fit_predict(dataset)
#     plt.scatter(X[:,0],X[:,1], c=cluster.labels_, cmap='rainbow')
    ax.scatter(dataset['x0'], dataset['x1'], c=y_agc, s=50, cmap='viridis')
    ax.set_title(title)
```

```python
# para a comparação escolhi o numero de clusters = 3 em ambos os modelos


fig, axList = plt.subplots(1, 2)
fig.set_size_inches(15, 5)
plot_kmean_cluster(kmeans_models[2], 'Kmeans 3 clusters', axList[0])
plot_agc_cluster(agc_models['ward'], 'AgglomerativeClustering 3 clusters ward', axList[1])
```

```python
# carregar a imagem

img = io.imread('bird_small.png')
io.imshow(img)
io.show()
```

```python
# image dimensions

rows = img.shape[0]
cols = img.shape[1]
colors = img.shape[2]
img.shape
```

```python
# create and fit model
data_img = img.reshape(rows * cols, colors)  
kmeans_model = KMeans(n_clusters=16)
kmeans_model.fit(data_img)
pd.DataFrame(data_img, columns=["r", "g", "b"]).head()
```

```python
# compress the image converting color

cmp_img = np.array([list(kmeans_model.cluster_centers_[label]) for label in kmeans_model.labels_])
cmp_img = np.clip(cmp_img.astype('uint8'), 0, 255)
pd.DataFrame(cmp_img, columns=["r", "g", "b"]).head()
```

```python
# resize to the original size

cmp_img_c = np.clip(cmp_img.astype('uint8'), 0, 255)
cmp_img_c = cmp_img_c.reshape(rows, cols, 3)
```

```python
# show and save the resulted img
io.imsave('cmp_img.png', cmp_img_c)
io.imshow(cmp_img_c)
io.show()
```
